## 作業目標

根據範例 ，完成以下問題：

* ① 取出今天所有的發文
* ② 如果想要依照類別分類，怎麼儲存會比較好？
* ③ 哪一個類別的文章最多




```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
print(soup.prettify())
```

### ① 取出今天所有的發文

```python
'''
Your Code
'''
for d in soup.find(class_="part_list_2").find_all("h3"):
    print(d.find(class_="date").text, d.find_all("a")[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
d.find(class_="tag").text
```

```python
'''
Your Code
'''

res = {}
for d in soup.find(class_="part_list_2").find_all("h3"):
    date = d.find(class_="date").text
    title = d.find_all("a")[-1].text
    tag = d.find(class_="tag").text
    res.setdefault(tag, [])
    res[tag].append({
        "title": title,
        "date": date
    })

print(res)
```

### ③ 哪一個類別的文章最多

```python
'''
Your Code
'''
Count = []
for i, tag in enumerate(res.keys()):
    Count.append([tag, len(res[tag])])
print(Count)

```
