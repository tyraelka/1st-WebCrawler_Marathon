## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
print(soup.prettify())
```

```python
for d in soup.find_all(class_="title"):
    print(d.text.replace("\t", "").replace("\n",""))
    try:
        r = BeautifulSoup(requests.get("https://www.ptt.cc"+d.find("a")["href"]).text, "html5lib")
        print("作者: "+ r.find(class_="article-meta-value").text)
    except:
        continue
```

```python
title = soup.find(class_="title").text.replace("\t", "").replace("\n", "")
title
```

```python
article = BeautifulSoup(requests.get("https://www.ptt.cc"+soup.find(class_="title").find("a")["href"]).text, "html5lib") #.find(class_="artical-meta-value").text
author = article.find_all(class_="article-meta-value")[0].text
author
```

```python
time = article.find_all(class_="article-meta-value")[-1].text
time
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
print("作者: "+author)
print("標題: "+soup.find(class_="title").text.replace("\t","").replace("\n",""))
print("時間: "+time)
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
'''
Your Code
'''
for d in soup.find_all(class_="title"):
    title = d.text.replace("\t", "").replace("\n", "")
    print("標題: "+title)
    try:
        r1 = BeautifulSoup(requests.get("https://www.ptt.cc"+ \
                                        d.find("a")["href"]).text, "html5lib")
        author = r1.find_all(class_="article-meta-value")[0].text
        time = r1.find_all(class_="article-meta-value")[-1].text
        print("作者: "+author)
        print("時間: "+time)
        print("-"*50)
    except:
        print("-"*50)
        continue
```
