## 作業目標

* 根據範例提供的 API ，完成以下問題：

    1. 取出知乎問題發問時間
    2. 取出第一筆與最後一筆回答的時間


```python
import requests, json

headers = {"user-agent": "my-app/0.0.1"}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',
                 headers = headers)
response = r.text

response
```

```python
data = json.loads(response)
info = []

for d in data["data"]:
    info.append(d)
info
```

```python
info[1]
```

### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

'''
Your Code
'''
for i in range(len(info)):
    print("第 %d 題問題發問時間: " %(i+1), info[i]["question"]["created"])
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間

'''
Your Code
'''
for i in range(len(info)):
    print("第 %d 題問題的第一筆回答時間為 %s" %((i+1), str(info[i]["question"]["updated_time"])))
    print("第 %d 題問題的最後一筆回答時間為 %s" %((i+1), str(info[i]["updated_time"])))
    print("-"*50)
```
