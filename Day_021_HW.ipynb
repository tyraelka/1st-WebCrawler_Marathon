## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import time
from datetime import datetime, timedelta
```

```python
url = "https://www.ettoday.net/news/news-list.htm"

#打開瀏覽器
browser = webdriver.Chrome("./chromedriver.exe")
browser.get(url)
```

```python
#取得昨天時間
yesterday = datetime.now() - timedelta(days=1)
yesterday = yesterday.strftime("%Y/%m/%d")
print("Yesterday: ", yesterday)
```

```python
#每2秒自動往下滑

js = "window.scrollTo(0, 100000);"

for i in range(30):
    time.sleep(2)
    browser.execute_script(js)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "lxml")

for d in soup.find(class_="part_list_2").find_all('h3'):
    if yesterday in d.find(class_="date").text :
        break
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
news = []

for d in soup.find(class_="part_list_2").find_all("h3"):
    if yesterday in d.find(class_="date").text:
        break
    news.append({
        "Time": d.find(class_="date").text,
        "Title": d.find_all("a")[-1].text
    })
```

```python
import pandas as pd
pd.DataFrame(news)
```

```python
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
two_hrs_time = datetime.now() - timedelta(hours=2)

news = []

for d in soup.find(class_="part_list_2").find_all("h3"):
    if datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M") < two_hrs_time:
        break
    
    news.append({
        "Time": d.find(class_="date").text,
        "Title": d.find_all("a")[-1].text
    })
pd.DataFrame(news)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
browser = webdriver.Chrome(executable_path="./chromedriver.exe")
browser.get(url)
```

```python
#取得三天前下午三點到五點的時間
date_time = datetime.now() - timedelta(3)
print("三天前日期: ", date_time)

date_time_month = date_time.strftime("%m")
date_time_day = date_time.strftime("%d")
print("三天前月份: ", date_time_month)
print("三天天日子: ", date_time_day)

three_day_15_clock = datetime.strptime(date_time.strftime("%Y/%m/%d") + " 15", "%Y/%m/%d %H")
three_day_17_clock = datetime.strptime(date_time.strftime("%Y/%m/%d") + " 17", "%Y/%m/%d %H")
print("三天前下午三點: ", three_day_15_clock)
print("三天前下午五點: ", three_day_17_clock)
```

```python
#選擇且點選日期
from selenium.webdriver.support.ui import Select

# selM = Select(browser.find_element_by_id("selM"))
# selM.select_by_value(date_time_month)
selD = Select(browser.find_element_by_id("selD"))
selD.select_by_value(date_time_day)
browser.find_element_by_id("button").click()
```

```python
import time
while True:
    browser.execute_script("window.scrollTo(0, 100000);")
    time.sleep(2)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    new_date = soup.find_all(class_="part_list_2")[-1].find_all("h3")[-1].find(class_="date").text
    print(new_date)
    if datetime.strptime(new_date, "%Y/%m/%d %H:%M") < three_day_15_clock:
        print("===STOP===")
        break        
```

```python
#取得資料
news = []

for d in soup.find(class_="part_list_2").find_all("h3"):
    #比兩小時早的話就結束
    if datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M") < three_day_15_clock:
        print("===STOP===")
        break
    
    if datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M") < three_day_17_clock:
        news.append({
            "Time": d.find(class_="date").text,
            "Title": d.find_all("a")[-1].text
        })
import pandas as pd
pd.DataFrame(news)
```
